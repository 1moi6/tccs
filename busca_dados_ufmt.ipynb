```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Função para processar uma página e extrair os dados da tabela
def process_page(url):
    response = requests.get(url)
    response.raise_for_status()  # Garante que erros HTTP sejam tratados
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontra a tabela no HTML (ajuste o seletor conforme necessário)
    table = soup.find('table')
    if not table:
        return []  # Retorna vazio se a tabela não for encontrada

    rows = table.find_all('tr')
    data = []

    for row in rows:
        cols = row.find_all(['td', 'th'])  # Inclui cabeçalhos na primeira iteração
        cols = [col.get_text(strip=True) for col in cols]
        if cols:
            data.append(cols)

    return data

# URL base e número de páginas
base_url = "https://sistemas.ufmt.br/sgpp/Publico?page="
num_pages = 465

# Lista para armazenar os dados
all_data = []
header = []

for page in range(1, num_pages + 1):
    url = f"{base_url}{page}"
    page_data = process_page(url)

    if page_data:
        if not header:  # A primeira página contém o cabeçalho
            header = page_data[0]
        all_data.extend(page_data[1:])  # Exclui cabeçalho nas páginas seguintes

    print(f"Página {page}/{num_pages} processada.")

# Cria um DataFrame
if all_data:
    df = pd.DataFrame(all_data, columns=header)
    print("Tabela consolidada criada com sucesso.")
else:
    print("Nenhuma tabela foi encontrada.")

# Salva o DataFrame em um arquivo CSV
df.to_csv("tabela_ufmt.csv", index=False)
print("Tabela salva em 'tabela_ufmt.csv'.")
```

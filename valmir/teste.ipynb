```python
import tifffile as tiff
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carregar a imagem TIFF
imagem = tiff.imread("L8_OLI_2013.tif")

# Verificar dimens√µes
print(f"Dimens√µes da imagem: {imagem.shape}")  # (Altura, Largura, Canais)

rgb = imagem[:, :, [3, 2, 1]].astype(np.float32)

# Normalizar para o intervalo [0,1]
rgb = (rgb - rgb.min()) / (rgb.max() - rgb.min())

# Exibir a imagem RGB corretamente normalizada
plt.figure(figsize=(8, 6))
plt.imshow(rgb)
plt.axis("off")
plt.title("Composi√ß√£o RGB Normalizada")
plt.show()

rgb = imagem[:, :, [5, 4, 3]]  # Ajuste conforme necess√°rio

# Normalizar os valores (se necess√°rio)
rgb = (rgb - rgb.min()) / (rgb.max() - rgb.min())

# plt.imshow((rgb<0.20).astype(float))
plt.imshow(rgb)  # Exibe apenas a primeira banda
plt.axis("off")
plt.title("Primeira Banda da Imagem TIFF")
plt.show()
```

```python

# Carregar a imagem TIFF
imagem_24 = tiff.imread("L8_OLI_2024.tif")

# Verificar dimens√µes
print(f"Dimens√µes da imagem: {imagem_24.shape}")

# Selecionar as bandas relevantes
nir = imagem_24[:, :, 4].astype(np.float32)  # Banda 5 (NIR)
red = imagem_24[:, :, 3].astype(np.float32)  # Banda 4 (Red)

# Evitar divis√£o por zero adicionando uma pequena constante
ndvi_24 = (nir - red) / (nir + red + 1e-10)

# Exibir a imagem do NDVI
plt.figure(figsize=(8, 6))
plt.imshow((ndvi_24<0.15).astype(float), vmin=-1, vmax=1)  # Cores: vermelho (-1) a verde (+1)
plt.colorbar(label="NDVI")
plt.axis("off")
plt.title("√çndice de Vegeta√ß√£o por Diferen√ßa Normalizada (NDVI)")
plt.show()
```

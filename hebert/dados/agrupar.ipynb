```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift
from sklearn.preprocessing import MinMaxScaler
```

```python
dados = pd.read_csv("nbb.csv")
```

```python
dados = pd.read_csv("nbb.csv")
columns = []
names = []
for cl in dados.columns:
    cln = cl.split(".")[0]
    if cln not in columns:
        names.append("_".join(cln.split(" ")).lower())
        columns.append(cl) 

dados = dados[columns]
dados.columns = names
dados = dados.dropna()
dados["total_min"] = dados["jogo"]*dados["min_em_quadra"]
dados
```

```python
dados_sort = dados.sort_values(by="min_em_quadra",ascending=False)
dados_sort = dados_sort[dados_sort["min_em_quadra"]>=17]
escala = MinMaxScaler()
X = escala.fit_transform(dados_sort.drop("jogadores",axis=1))
```

```python
agrupamento = KMeans(n_clusters=5,n_init=100)
dados_agrupados = agrupamento.fit_predict(X)
```

```python
rr = pd.DataFrame({'jogadores':dados_sort["jogadores"],'grupo':dados_agrupados})
```

```python
rr.to_csv("resultado.csv",index=False)
```

```python
dados = pd.read_csv("dadosatualizados.csv")
columns = []
names = []
for cl in dados.columns:
    cln = cl.split(".")[0]
    if cln not in columns:
        names.append("_".join(cln.split(" ")).lower())
        columns.append(cl) 

dados = dados[columns]
dados.columns = names
dados = dados.dropna()
dados
```

```python
dados.columns
```

```python
escala = MinMaxScaler()
X = escala.fit_transform(dados.drop(["jogador",'posição_em_quadra', 'jogo', 'min_em_quadra',
       'centesimos_de_minutos'],axis=1))
```

```python
agrupamento = KMeans(n_clusters=5,n_init=100)
dados_agrupados = agrupamento.fit_predict(X)
dados_agrupados
```

```python
dados["grupo"] = list(dados_agrupados)
```

```python
pd.crosstab(dados['grupo'], dados['posição_em_quadra'])
```

```python
modelo2 = AgglomerativeClustering(n_clusters=5)
grupamento = modelo2.fit_predict(X)
grupamento
```

```python
dados["grupo2"] = list(grupamento)
```

```python
pd.crosstab(dados['grupo2'], dados['posição_em_quadra'])
```

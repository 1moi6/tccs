```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift
from sklearn.preprocessing import MinMaxScaler
```

```python
dados = pd.read_csv("nbb.csv")
```

```python
dados = pd.read_csv("nbb.csv")
columns = []
names = []
for cl in dados.columns:
    cln = cl.split(".")[0]
    if cln not in columns:
        names.append("_".join(cln.split(" ")).lower())
        columns.append(cl) 

dados = dados[columns]
dados.columns = names
dados = dados.dropna()
dados["total_min"] = dados["jogo"]*dados["min_em_quadra"]
dados
```

```python
dados_sort = dados.sort_values(by="total_min",ascending=False)
```

```python
dados_sort
```

```python
dados.to_csv("nbb2.csv")
```

```python
dados_sort = dados_sort[dados_sort["total_min"]>=100]
escala = MinMaxScaler()
X = escala.fit_transform(dados_sort.drop("jogadores",axis=1))
```

```python
agrupamento = KMeans(n_clusters=5,n_init=10)
dados_agrupados = agrupamento.fit_predict(X)
```

```python
pd.DataFrame({'jogadores':dados_sort["jogadores"],'grupo':dados_agrupados})
```
